```python
from langchain_community.utilities import GoogleSerperAPIWrapper
from langchain_openai import OpenAI
from langchain.agents import initialize_agent, Tool
from langchain.agents import AgentType


llm = OpenAI(temperature=0)
search = GoogleSerperAPIWrapper()
tools = [
    Tool(
        name="Intermediate Answer",
        func=search.run,
        description="useful for when you need to ask with search"
    )
]

self_ask_with_search = initialize_agent(tools, llm, agent=AgentType.SELF_ASK_WITH_SEARCH, verbose=True)
self_ask_with_search.run("What is the hometown of the reigning men's U.S. Open champion?")
```

```python
from langchain_community.utilities import GoogleSerperAPIWrapper
from langchain_openai import OpenAI
from langchain.agents import initialize_agent, Tool
from langchain.agents import AgentType
from langchain_community.tools.google_finance import GoogleFinanceQueryRun
from langchain_community.utilities.google_finance import GoogleFinanceAPIWrapper

llm = OpenAI(temperature=0)
general_search = GoogleSerperAPIWrapper()
#tool = GoogleFinanceQueryRun(api_wrapper=GoogleFinanceAPIWrapper())
finance_search = GoogleFinanceQueryRun(api_wrapper=GoogleFinanceAPIWrapper())
tools = [
    Tool(
        name="General Search",
        func=general_search.run,
        description="useful for when you need to ask with search"
    ),
    Tool(
        name="Finance Search",
        func=finance_search,
        description="useful for when you need to ask with search for finance related queries"
    )
]

self_ask_with_search = initialize_agent(tools, llm, agent=None, verbose=True)
self_ask_with_search.run("What is the hometown of the Albert Einstein?")
#self_ask_with_search.run("What is the current share price of Apple?")
```
